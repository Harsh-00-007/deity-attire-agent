# 🎨 VastraGen AI Studio - Cloud GPU Deployment

This notebook is a 1-click deployment script to run the **Deity Attire Agent** on Google Colab's ephemeral NVIDIA T4 GPUs. 

It will:
1. Clone the repository
2. Install dependencies securely
3. Prompt for API keys (Groq & SerpApi)
4. Establish a secure Cloudflare tunnel to serve the Streamlit UI to the public web.

```python
# 1. Clone Repo & Install
!git clone https://github.com/Harsh-00-007/deity-attire-agent.git
%cd deity-attire-agent
!pip install -r requirements.txt langchain-groq -q

# 2. Get API Keys
import os
from getpass import getpass
os.environ["GROQ_API_KEY"] = getpass("Enter Groq API Key: ")
os.environ["SERPAPI_API_KEY"] = getpass("Enter SerpApi Key: ")

# 3. Download Cloudflare Tunnel (Enterprise-grade, no warning screens)
!wget -q -c -nc https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64
!chmod +x cloudflared-linux-amd64

# 4. Boot Streamlit in the background
!streamlit run app.py &>/content/logs.txt &

# 5. Start Cloudflare Tunnel and extract the secure URL
import time
print("\nBooting up Vision Engine & Cloudflare Tunnel...")
time.sleep(15) # Give Streamlit a few seconds to start

!nohup ./cloudflared-linux-amd64 tunnel --url http://localhost:8501 &> tunnel.log &
time.sleep(15)

print("\n=== YOUR APP IS LIVE AT ===")
!grep -o 'https://.*\.trycloudflare.com' tunnel.log
print("===========================\n")
```
